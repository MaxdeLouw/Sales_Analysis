This SQL notebook is just for practise and verification of the Pandas operation.

```python
import sqlite3
import pandas as pd

conn = sqlite3.connect(r"../Data/sales.db")

df = pd.read_csv(r"../Data/clean_sales_data.csv", encoding="latin1")

df.to_sql(
    "sales",
    conn,
    if_exists="replace",
    index=False
)

conn.close()
```

```python
query = """
SELECT 
    State,
    SUM(Sales) AS TotalSales,
    SUM(Profit) AS TotalProfit,
    ROUND(
        SUM(Profit) * 100.0 / SUM(Sales), 2
    ) AS ProfitMargin
FROM sales
GROUP BY State
ORDER BY TotalSales DESC;
"""

conn = sqlite3.connect(r"../Data/sales.db")

pd.read_sql(query, conn)
```
