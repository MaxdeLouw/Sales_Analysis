```python
import pandas as pd
```

The CSV with out sales data is imported from Kaggle. Below I'll import it, get a slice, then output the basic metrics.

```python
df = pd.read_csv(r"../Data/data_raw.csv", encoding="latin1")
df.head();
df.info();
df.describe();

```

```python
print(df.isnull().sum());
print(df.duplicated().sum());
```

First conlusions: 21 columns, 10k rows. No empty rows or duplicates. I'll group some columns to get early insights into Sales & profit per location and product category.

```python
df_location = df.groupby('State')[['Sales','Profit']].sum().sort_values(by='Sales', ascending=False).copy()
df_location['% profit'] = (df_location['Profit'] / df_location['Sales']) * 100
df_location.head(10)
```

Where outliers are interesting to explore:

```python
top_sales = df_location.nlargest(5, 'Sales')
top_profit = df_location.nlargest(5, 'Profit')
bottom_profit = df_location.nsmallest(5, 'Profit')
biggest_performer = df_location.nlargest(5, '% profit')
biggest_underperformer = df_location.nsmallest(5, '% profit')

print(top_sales)
print(top_profit)
print(bottom_profit)
print(biggest_performer)
print(biggest_underperformer)
```

```python
df_category = df.groupby('Sub-Category')[['Sales', 'Profit']].sum().sort_values(by='Sales', ascending=False).copy()
df_category['% profit'] = (df_category['Profit'] / df_category['Sales']) * 100
df_category.head(10)
```

Again looking for outliers:

```python
top_sales = df_category.nlargest(5, 'Sales')
top_profit = df_category.nlargest(5, 'Profit')
bottom_profit = df_category.nsmallest(5, 'Profit')
biggest_performer = df_category.nlargest(5, '% profit')
biggest_underperformer = df_category.nsmallest(5, '% profit')

print(top_sales)
print(top_profit)
print(bottom_profit)
print(biggest_performer)
print(biggest_underperformer)
```

Now we got some basic info on sales per location and category. I now want to analyse the sales and profit over time. Im interested in the data per quarter per year.

```python
df_extended = df.copy()
df_extended['Year'] = pd.to_datetime(df_extended['Order Date']).dt.year
df_extended['Quarter'] = pd.to_datetime(df_extended['Order Date']).dt.quarter
df_extended.groupby(['Year', 'Quarter'])[['Sales', 'Profit']].sum().sort_values(by='Year', ascending=True).head(10)
```

And then finally also a percentage growth added as a column. To visualise it later we will make a new dataframe out of this.

```python
df_trend = df_extended.groupby(['Year', 'Quarter'])[['Sales', 'Profit']].sum().sort_values(by='Year', ascending=True).copy()
df_trend['Sales Growth'] = df_trend['Sales'].pct_change() * 100
df_trend['Profit Growth'] = df_trend['Profit'].pct_change() * 100
df_trend
```

Here we see some periodicity to sales YoY. There is a suggested sales drop in Q1 of every year.  YoY analysis, by comparing the quarters, could confirm this.

```python
df_trend['Sales_Growth_YoY'] = df_trend['Sales'].pct_change(periods=4) * 100
df_trend['Profit_Growth_YoY'] = df_trend['Profit'].pct_change(periods=4) * 100
df_trend
```

Indeed a recurring seasonal pattern is visible in the data. YoY remains positive for most quarters, indicating that the business does continue to grow despite these drops.

This is it for the EDA (exploratory data analysis) part. Last step is to convert the created dataframes back to CSV

```python
df_location.to_csv("../Data/df_location_summary.csv")
df_category.to_csv("../Data/df_category_summary.csv")
df_trend.to_csv("../Data/df_trend_summary.csv")
df_extended.to_csv("../Data/clean_sales_data.csv", index=False)
```
